## 4. 인코더 모델

* Transformer 모델의 인코더 모듈만 사용
* 각 단계에서 어텐션 계층이 초기/원본 입력 문장의 모든 단어에 액세스
* 주로 "양방향" attention
* 일반적으로 초기 문장을 손상시킨 뒤(예: masking) 손상시킨 문장을 다시 복원하는 과정을 통해 모델 학습 진행

* 태스크
    1. 문장 분류
    2. 개체명 인식
    3. 단어 분류
    4. 추출형 질의 응답

    등 전체 문장에 대한 이해가 필요한 작업

* 예)
    1. ALBERT
    2. BERT
    3. DistillBERT
    4. ELECTRA
    5. RoBERTa
